```python
from scipy.stats import norm
from math import sqrt
```

```python
def one_sided_hypo (mean_train,mean,sd,sample_size,alpha):
    actual_z=(norm.ppf(alpha))
    hypo_z=(mean_train-mean)/(sd/sqrt(sample_size))
    print('Actual z value:',actual_z)
    print('hypothesis z value:', hypo_z,'\n')
    if(hypo_z>=actual_z):
        return True
    else:
        return False
sample_size=100
mean_train=4.15
mean=4.18
sd=0.526
alpha=0.05
print('H0 : M <=',mean)
print('H1 : M >',mean)
print('aplha value is :',alpha,'\n')
reject = one_sided_hypo(mean_train,mean,sd,sample_size,alpha)
if(reject):
    print("Null Hypothesis is  rejected")
else:
    print("Failed to reject Null Hypothesis")
```
